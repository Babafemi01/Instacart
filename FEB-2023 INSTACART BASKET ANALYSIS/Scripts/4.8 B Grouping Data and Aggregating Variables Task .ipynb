# Aggregating data with transform() to create a flag column.¶
Flags help you target specific groups within your data according to a set of conditions, allowing for segmented analysis

# creating a flag for customers with many orders, after which have set my criteria. To achieve the above; 3 steps to get the result are 
1- Split the data into groups based on the “user_id” column
2- Apply the transform() function on the “order_number” column to generate the maximum orders for each user
3- Create a new column, “max_order,” into which you’ll place the results of your aggregation

```python
# the 3 steps are in the code;new column max_order, groupby the user_id and transform order_no with the max() in Np Library 
df_new_products['max_order'] = df_new_products.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# to view the newly created column max_order
df_new_products.head()
```

## Deriving columns with loc(). Now you have created the new column you wanted.
Next is to set that criteria to perform your segmented analysis ( which is to see which customer using the max_order new column
has a regular customer status, loyal customer status and new customer status.

```python
# setting the criteria that a new column loyalty_flag created should show a string value of 'loyal customer' if the 
max_order for that row is > 40 within my df
df_new_products.loc[df_new_products['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# setting the criteria that a new column loyalty_flag created should show a string value of 'Regular customer' if the 
max_order for that row is <= 40 and > 10 within my df
df_new_products.loc[(df_new_products['max_order'] <= 40) & (df_new_products['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# setting the criteria that a new column loyalty_flag created should show a string value of 'New customer' if the 
max_order for that row is < 10 within my df
df_new_products.loc[df_new_products['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

#  Create a spending flag for each user based on the average price across all their orders using the following criteria:
Using the following criteria : If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” 2) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.

I will use the 3 steps below:
1- Split the data into groups based on the “user_id” column
2- Apply the transform() function on the “order_number” column to generate the maximum orders for each user
3- Create a new column, “max_order,” into which you’ll place the results of your aggregation

```python
# looking at the descriptive statistics of the price column
df_new_products['prices'].describe()
```

```python
# the 3 steps are in the code;new column average_price, groupby the user_id and transform prices with the max() in Np Library 
df_new_products['average_price'] = df_new_products.groupby(['user_id'])['prices'].transform(np.max)
```

```python
# to view the newly created max_price column
df_new_products.head()
```

## Deriving columns with loc(). Now you have created the new column you wanted.
Next is to set that criteria to perform your segmented analysis ( which is to see which customer using the average_price new column
has a high spender or low spender status

criteria 1) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
criteria 2) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# setting the criteria that a new column spending_flag created should show a string value of 'low spender' if the 
average_price for that row is < 10 within my df
df_new_products.loc[df_new_products['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# setting the criteria that a new column spending_flag created should show a string value of 'High spender' if the 
average_price for that row is >= 10 within my df
df_new_products.loc[df_new_products['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_new_products['spending_flag'].value_counts(dropna = False)
```

# Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

Using the following criteria "If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
"If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
"If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# looking at the descriptive statistics of the days_since_prior_order column. The median is 7 days
df_new_products['days_since_prior_order'].describe()
```

```python
# the 3 steps are in the code;new column order_freq, groupby the user_id and transform prices with the median() in Np Library
df_new_products['order_freq'] = df_new_products.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# to view the newly created order_freq column
df_new_products.head()
```

## Deriving columns with loc(). Now you have created the new column you wanted.
Next is to set that criteria to perform your segmented analysis ( which is to see which customer using the order_freq new column
has frequent , non-frequent and regular customer.

criteria 1) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.
criteria 2) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.
criteria 3) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer

```python
# setting the criteria that a new column order_freq_flag created should show a string value of 'Non-frequent customer' if the 
order_freq (median of days_since_prior_order) for that row is > 20 within my df
df_new_products.loc[df_new_products['order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
# setting the criteria that a new column order_freq_flag created should show a string value of 'Regular customer' if the 
order_freq for that row is > 10 and <= 20 within my df
df_new_products.loc[(df_new_products['order_freq'] > 10) & (df_new_products['order_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
# setting the criteria that a new column order_freq_flag created should show a string value of 'frequent customer' if the 
order_freq for that row is <= 10 within my df
df_new_products.loc[df_new_products['order_freq'] <= 10, 'order_freq_flag'] = 'frequent customer'
```
