# Using the user defined fx, where you will actually input the IF and THEN in the statement

```python
# defining my def and naming it as price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# after defining my fx price_label, then i can use it within a df(newly creattd subset df) as a filter 
(conditional statements)
df_new_products_merged_orders['price_range'] = df_new_products_merged_orders.apply(price_label, axis=1)
```

```python
# counting the number of rows within the 1 million rows(that is d newly created subset) that was affected by the IF & Then
statement that was defined above
df_new_products_merged_orders['price_range_loc'].value_counts(dropna = False)
```

# Stakeholders want to know how busy each days of the week is, so they can use such info to tailor ads on specific days.

```python
# first you need to know on which days most orders take place in my df_new_products df. (count the values)
df_new_products['order_dow'].value_counts(dropna = False)
```

```python
# using the above info to create a new column to show the string that contain only 3 values (busiest day,least busy & regular day)
result = []

for value in df_new_products["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# running your new column 'result' whic shows how it has loop all thru the rows in that new column
result
```

# combine the 'result' with your df 'df_new_products' THEN do a count to summarise those new values 
(busiest day, least busy, regular busy) within the new column RESULT
By adding the values in result to a new column in your dataframe, you can use your new labels more effectively, 
for instance, by being able to see at a glance how many sales occur on each of the different types of days.

```python
# combine the 'result' with your df 'df_new_products'
df_new_products['Busiest day'] = result
```

```python
# Then make a count , so as to the see the summarized values.
df_new_products['Busiest day'].value_counts(dropna = False)
```

#, EXERCISE 4.7 Now the Stakeholders, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method

```python
# From above, i already know that Sunday and Saturday are the 2 busiest days. Now i will use string Busiest day, slowest day.making use of the days of the week 0-6

Days = []

for value in df_new_products["order_dow"]:
  if value == 0 or value == 1:
    Days.append("Busiest days")
  elif value == 4 or value == 3:
    Days.append("Slowest days")
  else:
    Days.append("Regular days")
```

```python
# running your new column 'Days' whic shows how it has loop all thru the rows in that new column
Days
```

```python
# combine the 'column Days' with your df 'df_new_products'
df_new_products['Busiest days'] = Days
```

```python
# Then make a count , so as to the see the summarized values meaning the frequency count
df_new_products['Busiest days'].value_counts(dropna = False)
```

# The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day

```python
# first you need to know on which hours most orders take place in my df_new_products df. (count the values)
df_new_products['order_hour_of_day'].value_counts(dropna = False)
```

## Observations: The 24 hours of the day will need to be turned to 3 values, with each having a string 'Most orders' (hour 10, 11, 14, 15, 13, 12, 16, 9 - from 9am to 4pm), 'Average orders' (hour 17, 8, 18, 19, 20, 7, 21, 22 - from 7am to 8am and from 5pm to 10pm), or 'Fewest orders' (hour 23, 6, 0, 1, 5, 2, 4, 3 - from 11pm to 6am)

```python
# using the above info to create a new column to show the string that contain only 3 values Most orders, Average Orders,Fewest orders
Am doing this on my subset df: df_new_products_merged_orders whc contain one million rows, because i was herring space error.
hours = []

for hour in df_new_products_merged_orders["order_hour_of_day"]:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# create the 'busiest_period_of_day' column within my newly created subset df
df_new_products_merged_orders['busiest_period_of_day'] = hours
```

```python
# checking the freq of the three values that constitute the busiest_period_of_day within my original df. df_new_products
df_new_products['busiest_period_of_day'].value_counts(dropna = False)
```
