```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}
```

```python
# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

```python
#Convert the dictionary into dataframe
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
```

```python
#Convert the dictionary into dataframe
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

# Merging 2 or more data sets using the merge fx

```python
# merge the df and df_2
df_merged = df.merge(df_2, on = 'customer_id')
```

```python
df_merged
```

```python
# Using multiple columns name on the (on) argument, like 'customer_id' and 'month'
df_merged = df.merge(df_2, on = ['customer_id', 'month'])
```

```python
df_merged
```

```python
# Using the INDICATOR to quickly and easily see if my key parameters both exist btw the 2 df or exist in 1 df.
df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
df_merged
```

```python
# to count the values in the new column added '_merge' due to the INDICATOR argument
df_merged['_merge'].value_counts()
```

```python
# Test merge without overwriting
pd.merge(df,df_2, on = ['customer_id', 'month'], indicator = True)
```

# Using the 'HOW' argument

```python
# Specifying how you want your dataframes 2be joined by using 'how' argument. how = 'outer', how = 'inner' or left or right
pd.merge(df,df_2, on = ['customer_id', 'month'], how = 'inner')
```

```python
# Outer Join
pd.merge(df,df_2, on = ['customer_id', 'month'], how = 'outer')
```
